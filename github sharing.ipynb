```python
import pandas as pd
import numpy as np
import folium
```

```python
Toranto_n= pd.read_csv('Toranto_neighbourhood.csv',index_col= 'PostalCode')
```

```python
Toranto_neighbour=Toranto_n.replace('Not assigned',np.nan)
Toranto_neighbour.dropna(axis=0, inplace=True)
```

```python
Toranto_neighbour.reset_index()
```

## Question 2

```python
geo_coordinate = pd.read_csv('Geospatial_Coordinates.csv', index_col='Postal Code') 
```

Geo coordinate dataframe

```python
geo_coordinate
```

Jion neighbourhood dataframe and geo coordinate data

```python
Toranto_neighbourhood = Toranto_neighbour.join(geo_coordinate)
```

```python
Toranto_neighbourhood.set_index('index')
```

## Question number 3

Plotting neighbourhood

```python
latitude= 43.6532
longitude=-79.3832
radius= 30000
```

```python
# creating map zoom at present Location
map = folium.Map(location=[latitude,longitude],zoom_start=12)

# adding the current location to map
folium.CircleMarker(location=[latitude,longitude],radius=5,color='red',fill=True,fill_color='red').add_to(map)
folium.Marker(location=[latitude,longitude],popup="<strong> Current Location</strong>",icon = folium.Icon(color='red')).add_to(map)

# drawing radius of search
folium.Circle(location=[latitude,longitude],radius=radius,color='blue',opacity=0.5,fill=True,fill_color='blue').add_to(map)

# adding all the popular places around
for lat,lon,borough in zip(Toranto_neighbourhood.Latitude,Toranto_neighbourhood.Longitude,Toranto_neighbourhood.Borough):
  folium.CircleMarker(location=[lat,lon],radius=5,color="red",fill=True,fill_color='red').add_to(map)
  folium.Marker(location=[lat,lon],
              popup= folium.Popup(('<strong><font color ="red">borough :     </font>'+borough+'</strong><br>'
                                  
                                   ),max_width=250)).add_to(map)

map
```
